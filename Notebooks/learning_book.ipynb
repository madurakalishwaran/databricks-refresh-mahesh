```python
from pyspark.sql import SparkSession, functions as F
spark = SparkSession.builder.appName("Spark DataFrames").getOrCreate()
delta = spark.read.table("chocolate_sales")
```

```python
delta.printSchema()
delta.select(F.col("Shipdate").cast("date")).printSchema()
delta.withColumn("Shipdate", F.to_date(F.col("Shipdate"))).printSchema()
delta.printSchema()
```
