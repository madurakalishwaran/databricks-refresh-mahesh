```python
from pyspark.sql import SparkSession, functions as F
spark = SparkSession.builder.appName("Learning").getOrCreate()
```

```python
file_path = "/Volumes/workspace/default/my_datas/sample-chocolate-sales-data-all.csv"

df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv(file_path)
```

```python
df.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("default.chocolate_sales")
```

```python
drop_tables = ["chocolate_sale", "none", "sales"]
for t in drop_tables:
  try:
    spark.sql(f"DROP TABLE IF EXISTS {t}")
  except:
    pass
```

```python
r = spark.read.table("chocolate_sales")
r.groupby("GID").agg(
    F.count(F.lit(0)).alias("total_rows"),
    F.sum("Boxes").alias("total_boxes"),
    F.avg("Amount").alias("avg_amount")
).show()
```
